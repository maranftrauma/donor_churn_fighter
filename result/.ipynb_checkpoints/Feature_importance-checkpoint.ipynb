### Feature importance notebook

This notebook takes the best_model_feature_importance.csv generated by Best_Model.R script in which the feature importance for the best model was generated for the train_fold 31. 

The training process was made using library 'extratrees' but this package does not contain feature importance methods. To get feature importance Best_model.R uses the library 'caret' and train with method 'extratrees' with the best hyperparameters(according to model seleccion in 'Model_Selection_Large_Cohort.py'.

#### Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import pathlib
%matplotlib inline
```

#### Import CSV

```python
df = pd.read_csv('best_model_feature_importance.csv')
```

```python
df = df.rename(columns={"Unnamed: 0": "feature"}).sort_values(by = 'importance', ascending=True)
df = df.tail(20)
```

#### Plot

```python
# Set general setup
nrows = 1  # Number of rows
ncols = 1 # Number of columns
fontsize = 14 # General fontsize
     
fig, axis = plt.subplots(figsize = (12,6))

axis.spines['top'].set_visible(False)
axis.spines['right'].set_visible(False)
axis.set_xlabel('Relative importance')
axis.set_ylabel('20 top features')
axis.set_title('Feature importance')

axis.barh(df['feature'], df['importance'], 
          align='center',
          color = '#4daf4a')

fig.tight_layout()

# Save to disc
name_file = 'Best_Model_Feature_Importance' + '.png'
file_to_save = pathlib.Path.cwd().joinpath('images',f'{name_file}')
plt.savefig(file_to_save)
```
